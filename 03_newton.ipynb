## Newton - Raphson
Η μέθοδος Newton-Raphson είναι μια επαναληπτική αριθμητική μέθοδος για την εύρεση των ριζών μιας συνάρτησης με πραγματική αξία. Χρησιμοποιείται συνήθως για την επίλυση μη γραμμικών εξισώσεων και προβλημάτων βελτιστοποίησης.

### Example
Find the square root of a number using the Newton-Raphson method. We define the function `f(x)=x^2-4`, and the derivative function `f_prime(x)=2x`.
The `newton_raphson` function iteratively updates the guess for the root `x0` using the Newton-Raphson formula: x -= f(x) / f_prime(x). It continues iterating until the function value f_val becomes smaller than the tolerance tol, or until the maximum number of iterations max_iter is reached.

```python
import numpy as np

def newton_raphson(f, f_prime, x0, tol=1e-6, max_iter=100):
    """
    Newton-Raphson method for finding the root of a function.
    
    Parameters:
    - f: The function for which the root is sought.
    - f_prime: The derivative of the function.
    - x0: The initial guess for the root.
    - tol: Tolerance for convergence (default: 1e-6).
    - max_iter: Maximum number of iterations (default: 100).
    
    Returns:
    - The approximate root of the function.
    """
    x = x0
    for i in range(max_iter):
        f_val = f(x)
        if abs(f_val) < tol:
            return x
        f_prime_val = f_prime(x)
        x -= f_val / f_prime_val
    return x
```

```python
# Example: Finding the square root of a number using the Newton-Raphson method

def f(x):
    return x**2 - 4  # The function x^2 - 4 = 0

def f_prime(x):
    return 2 * x  # The derivative of x^2 - 4
```

```python
# Solve for the root using the Newton-Raphson method
root = newton_raphson(f, f_prime, x0=1.5)

print("Approximate root:", root)
print("Actual square root:", np.sqrt(4))
```

```python
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 4

def f_prime(x):
    return 2 * x

def newton_raphson(f, f_prime, x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        f_val = f(x)
        if abs(f_val) < tol:
            return x
        f_prime_val = f_prime(x)
        x -= f_val / f_prime_val
    return x

# Generate x values for plotting
x_vals = np.linspace(0, 3, 400)

# Compute function values
y_vals = f(x_vals)

# Find the root using the Newton-Raphson method
root = newton_raphson(f, f_prime, x0=1.5)

# Plot the function and the root
plt.plot(x_vals, y_vals, label='f(x) = x^2 - 4')
plt.scatter(root, 0, color='red', label='Approximate Root')

plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Newton-Raphson Method')
plt.legend()
plt.grid(True)
plt.show()
```

## :uadratic transformation to a function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original function
def original_func(x):
    return np.sin(x)

# Define the quadratic transformation
def quadratic_transform(x):
    return original_func(x)**2

# Generate x values for plotting
x_vals = np.linspace(-np.pi, np.pi, 100)

# Compute the original function values
y_original = original_func(x_vals)

# Compute the transformed function values
y_transformed = quadratic_transform(x_vals)

# Plot the original and transformed functions
plt.plot(x_vals, y_original, label='Original Function: sin(x)')
plt.plot(x_vals, y_transformed, label='Quadratic Transformation: sin^2(x)')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Quadratic Transformation')
plt.legend()
plt.grid(True)
plt.show()
```
